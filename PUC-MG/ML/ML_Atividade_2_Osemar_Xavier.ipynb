````python
# Abaixo carrego as principais bibliotecas para analisar o dataframe
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier #DecisionTreeRegressor 
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
````

````python
# Abaixo fiz o carregamento dos dados para a variável auto_df
ereader_df = pd.read_excel('Atividade 3 - Bases.xlsx')
ereader_df.head()
````

````python
# Descrição dos dados utilizando .info precisaremos 
# transformar os dados em número binário
ereader_df.info()
````

````python
# Os únicos valores numéricos são o ID_usuario e idade
ereader_df.describe()
````

````python
# Removi as colunas que não possuem características binárias 
ereader_df.drop(['ID_usuário', 'Idade', 'Adoção_eReader'], axis=1)
````

````python
#Selecionar todas colunas exceto a classe

X_dict = ereader_df.iloc[:,ereader_df.columns != 'Adoção_eReader'].T.to_dict().values()
# X_dict = clima_nominal.iloc[:,0:4].T.to_dict().values()

# Exibe o dado convertido em dicionario.
print("Atributos:", X_dict)
````

````python
# Nessa etapa transformei as variáveis categóricas
vect = DictVectorizer(sparse=False)
````

````python
X_train = vect.fit_transform(X_dict)

# Abaixo faço a exibição da estrutura do dado convertido em binário.
print("Shape do dado de treinamento apos binarizacao (dummies): {0}".format(X_train.shape))
X_train
````

````python
vect.feature_names_
````

````python
ereader_df.head(3)
````

````python
le = LabelEncoder() 

y_train = le.fit_transform(ereader_df['Adoção_eReader'])

print("Labels:", y_train)
````

````python
le.classes_
````

# **INDUÇÃO DO MODELO**

Os três passos para indução de um modelo são:

1.   Instanciar o modelo: ``` DecisionTreeClassifier()```
2.   Treinar o modelo: ```fit()```
3.   Testar o modelo: ```predict()```

````python
treeClima = DecisionTreeClassifier(random_state=0, criterion='entropy')

treeClima = treeClima.fit(X_train, y_train)
treeClima
````

````python
treeClima.predict(X_train)
````

````python
y_train
````

# **VISUALIÇÃO DAS DECISÕES**

````python
import pydotplus
from IPython.display import Image
# Create DOT data
dot_data = tree.export_graphviz(treeClima, out_file=None, 
                                feature_names=vect.feature_names_,  
                                rounded =True,
                                filled=True,
                                class_names=le.classes_)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
````
