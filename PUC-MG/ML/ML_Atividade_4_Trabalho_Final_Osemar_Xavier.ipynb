```python
# Contando os valores de NaN numa coluna com o .isna().sum() 
# Nesse caso foi observado 313 valores NaN no campo MINIMUM_PAYMENTS
# e 1 caso de NaN no campo CREDIT_LIMIT 
cred_df.isna().sum()
```

```python
# Na coluna CRED_LIMIT foi identificado 1 'NaN' 
# Fiz a substituição desse valor pelo valor da média da coluna
mean_value=cred_df['CREDIT_LIMIT'].mean()
cred_df['CREDIT_LIMIT']=cred_df['CREDIT_LIMIT'].fillna(mean_value)
```

```python
# Na coluna MINIMUM_PAYMENTS foi identificado 313 'NaN' 
# Fiz a substituição desse valor pelo valor da média da coluna
mean_value=cred_df['MINIMUM_PAYMENTS'].mean()
cred_df['MINIMUM_PAYMENTS']=cred_df['MINIMUM_PAYMENTS'].fillna(mean_value)
```

```python
# Fazendo novo somatório dos 'NaN' verificamos que não existem mais 'NaN' no dataset
cred_df.isna().sum()
```

```python
# Utilizei o método corr() para encontrar a matriz de 
# correlação utilizando o método Pearson.
a = cred_df.corr()
a
```

```python
# Visualmente fica complicado visualizar a correlação por números
# Utilizei o heatmap para exibir os valores de correlação no mapa de calor.
plt.figure(figsize=(20,10))
sns.heatmap(a,vmin=-1,vmax=1,center=0,annot=True)
```

```python
# Deletado o campo 'CUST_ID'
cred_df = cred_df.drop('CUST_ID', axis=1)
cred_df.head(3)
```

```python
# Traz os valores unicos da coluna "TENURE"
cred_df['TENURE'].unique()
```

```python
# Abaixo plotei as features em gráfico para entendimento do dataset 
for i in cred_df:
    sns.displot(x = i, data = cred_df)
    plt.title(i)
```

```python
# Cria um array com todos os valores do dataframe
X = cred_df.iloc[:,:].values
X
```

```python
# A biblioteca sklearn do Python nos oferece a função StandardScaler() 
# para padronizar os valores dos dados em um formato padrão. Em seguida
# fit_transform () é usado nos dados de treinamento para que possamos escalar 
# os dados de treinamento e também aprender os parâmetros de escalonamento 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

# MODELO
Clusterização usando Kmeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(random_state=42)
```

# GRÁFICO COTOVELO PARA VALOR K**

```python
from yellowbrick.cluster import KElbowVisualizer
methods = ['distortion', 'silhouette', 'calinski_harabasz']
for i in methods:
    print(i)
    visualizer = KElbowVisualizer(kmeans, k=(2,20), metric = i, timings = False)
    visualizer.fit(X)
    visualizer.show()
```

# RESULTADO
O melhor valor para K é 3, para esse modelo.

```python
km_sample = KMeans(n_clusters=3)
km_sample.fit(cred_df)
```

```python
labels_sample = km_sample.labels_
```

```python
cred_df['N_CLUSTER'] = labels_sample
```

```python
sns.countplot(cred_df["N_CLUSTER"])
```

```python
sns.set_palette('Set2')
sns.scatterplot(cred_df['BALANCE'],cred_df['PURCHASES'],hue=cred_df['N_CLUSTER'],palette='Set1')
```

# CONCLUSÃO FINAL

Usamos K Means para entender padrões com base em clusters no conjunto de dados dimensionado. Fiz a plotagem dos gráficos 'distortion', 'silhouette', 'calinski_harabasz' com o objetivo de encontrar o valor ideal de K. Mas como podemos ver os gráficos de dispersão após a aplicação do agrupamento (ótimo k = 3), os agrupamentos ainda não são tão distintos quanto se espera que fossem. Inicialmente, quando aplicamos o KMeans com todas as features do conjunto de dados, chegamos a um ponto em que não é possível ter uma definição clara desses clusters, mesmo com o valor K ideal. Abaixo iniciei novo estudo utilizando Dendograma, que é um diagrama de árvore freqüentemente usado para demonstrar a organização dos clusters produzidos por agrupamento hierárquico. Creio tem muitas oportunidades para evoluir em novas análises para melhorar o resultado da clusterização. 

```python
plt.figure(figsize=(15,15))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distances')
plt.show()
```
