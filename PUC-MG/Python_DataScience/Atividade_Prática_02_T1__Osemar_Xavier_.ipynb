4) Apresente 3 exemplos de aplicações de conjuntos

```python
planetas1 = {'Vênus', 'Mercúrio', 'Terra', 'Netuno', 'Marte'}
planetas2 = {'Terra', 'Júpiter', 'Urano', 'Saturno', 'Marte'}
```

```python
print(f'A uniao dos conjuntos 1 e 2, é igual a: {planetas1 | planetas2}')
print(f'A diferença entre os conjuntos 1 e 2, é igual a: {planetas1 - planetas2}')
print(f'A intersecção entre os conjuntos 1 e 2, é igual a: {planetas1 & planetas2}')
```
