Fazer a importação do arquivo doencas.csv. A base de dados traz informações de saúde de algumas pessoas.

Colunas da base de dados:

* HeartDisease: Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI)
* BMI: Body Mass Index (BMI)
* Smoking: Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes]
* AlcoholDrinking: Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week
* Stroke: (Ever told) (you had) a stroke?
* PhysicalHealth: Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30
* MentalHealth: Thinking about your mental health, for how many days during the past 30 days was your mental health not good?
* DiffWalking: Do you have serious difficulty walking or climbing stairs?
* Sex: Are you male or female?
* AgeCategory: Fourteen-level age category
* Race: Imputed race/ethnicity value
* Diabetic: (Ever told) (you had) diabetes?
* PhysicalActivity: Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
* GenHealth: Would you say that in general your health is...
* SleepTime: On average, how many hours of sleep do you get in a 24-hour period?
* Asthma: (Ever told) (you had) asthma?
* KidneyDisease: Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease?
* SkinCancer: (Ever told) (you had) skin cancer?

```python
# Carregando as bibliotecas que irei utilizar no tratamento do Dataframe
import pandas as pd
import numpy as np
```

1) Renomeie os nomes das colunas para a tradução em português.

```python
# Renomeando os nomes das colunas com a tradução
doencas_df.rename(columns={'HeartDisease':'Doenca_Cardio', 'BMI':'IMC', 
                            'Smoking':'Fumante', 'AlcoholDrinking':'Bebida_Alcoolica', 
                            'Stroke':'AVC', 'PhysicalHealth':'Saude_Fisica', 
                            'MentalHealth':'Saude_Mental', 'DiffWalking':'Dificuld_Andar', 
                            'Sex':'Sexo', 'AgeCategory':'Categ_Idade',
                            'Race':'Raca', 'Diabetic':'Diabetes', 
                            'PhysicalActivity':'Ativid_Fisica', 'GenHealth':'Saude_Geral', 
                            'SleepTime':'Tempo_Sono', 'Asthma':'Asma', 
                            'KidneyDisease':'Doenca_Renal', 'SkinCancer':'Cancer_Pele'}, inplace=True)

# Listar o dataframe com as novas colunas
doencas_df.head()
```

2) Faça uma análise descritiva da base de dados.

```python
doencas_df.describe().T
```

3) Qual o percentual de homens e mulheres na base?


```python
total = doencas_df.groupby('Sexo').agg(n = ('Sexo','count')).reset_index()
total['(%)'] = 100*total['n']/total['n'].sum()
total
```

4) Informar quantas pessoas tiveram alguma doença e/ou lesão física há mais de 5 dias.



```python
Mais_5d = doencas_df.loc[(doencas_df['Saude_Mental']>5) & (doencas_df['Saude_Fisica']> 5)]
print(f'O total de alguma doença e/ou lesão física há mais de 5 dias contabilizam {Mais_5d.shape[0]} casos.')
```

5) Calcular a mediana do indíce de massa corporal (BMI).

7) Determinar a faixa etária com mais ocorrência de câncer de pele.

8) Definir por sexo a quantidade de pessoas com e sem problemas cardíacos.


```python
total = cancer_pele_df.groupby(['Sexo','Doenca_Cardio']).agg(n=('Sexo', 'count')).reset_index()
total['(%)'] = 100*total['n']/total['n'].sum()
total
```

9) Associe doença cardíaca com alguma outra doença na base.

```python
condicao = [doencas_df['Doenca_Cardio'] == 'Yes',
            doencas_df['AVC'] == 'Yes']

resultados = ['Risco_AVC', 'Risco_Cardio']
doencas_df['Classificacao'] = np.select(condicao, resultados, 'Ok')

doencas_df.head()
```

10) Filtrar as pessoas que possuem mais de 3 tipos de doenças.

```python
doencas_df2 = new = doencas_df[['AVC', 'Dificuld_Andar', 'Diabetes','Asma','Doenca_Renal','Cancer_Pele']].copy()
doencas_df2 = doencas_df2.replace(to_replace = ['Yes','No','No, borderline diabetes','Yes (during pregnancy)'],value = ['1','0','0','1'])

doencas_df2 = doencas_df2.astype({'AVC':'int',
                                  'Dificuld_Andar':'int',
                                  'Diabetes':'int',
                                  'Asma':'int',
                                  'Doenca_Renal':'int',
                                  'Cancer_Pele':'int'})

doencas_df2['sum_doencas'] = doencas_df2['AVC'] + doencas_df2['Dificuld_Andar']  + doencas_df2['Asma'] + doencas_df2['Doenca_Renal'] + doencas_df2['Cancer_Pele']
```

```python
print(doencas_df2[doencas_df2['sum_doencas'] > 3])
```
