```python
import numpy as np
```

1) Solicite ao usuário que informe a quantidade de linhas (n) e colunas (m) e crie um ndarray nxm com valores aleatórios.

```python
n = int(input('Digite a dimensão n da matriz: '))
m = int(input('Digite a dimensão m da matriz: '))

x = np.matrix(np.arange(n*m).reshape((n, m)))
print(x)
```

2) Retorne a média dos valores de cada uma das n linhas

```python
print(x.mean(1))
```

3) Retorne a mediana dos valores das m colunas

```python
x = np.array([x])
median=np.median(x)
print(f'A mediana dos valores das m colunas é {median}.')
```

4) Retorne o valor máximo de todo array

```python
arr = np.array([x])
maxElement = np.amax(arr)
print(f'O máximo valor do array é {maxElement}.')
```

5) Faça uma função que receba um ndarray de valores inteiros e retorne um ndarray de mesma dimensão, onde os valores negativos serão substituídos por True e os valores positivos por False.

```python
print(np.isinf(arr))
```
